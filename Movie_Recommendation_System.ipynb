```python
movie_name = "Interstellar"

url = "https://api.themoviedb.org/3/search/movie"

params = {
    "api_key": API_KEY,
    "query": movie_name
}

response = requests.get(url, params=params)
data = response.json()

data
```

```python
first_movie = data["results"][0]

print(first_movie["title"])
print(first_movie["overview"])
print(first_movie["release_date"])
print(first_movie["vote_average"])
```

```python
movie_id = first_movie["id"]

print(movie_id)
```

```python
recommend_url = f"https://api.themoviedb.org/3/movie/{movie_id}/recommendations"

params = {
    "api_key": API_KEY
}

response = requests.get(recommend_url, params=params)
recommend_data = response.json()

recommend_data
```

```python
recommendations = recommend_data["results"]

for movie in recommendations[:10]:
    print(movie["title"])
```

```python
movies_list = []

# Add searched movie first
movies_list.append({
    "title": first_movie["title"],
    "overview": first_movie["overview"],
    "release_date": first_movie["release_date"],
    "rating": first_movie["vote_average"]
})

# Add TMDB recommended movies
for movie in recommendations[:10]:
    movies_list.append({
        "title": movie["title"],
        "overview": movie["overview"],
        "release_date": movie.get("release_date", ""),
        "rating": movie["vote_average"]
    })

df = pd.DataFrame(movies_list)

df
```

```python
df["features"] = (
    df["title"].fillna("") + " " +
    df["overview"].fillna("") + " " +
    df["release_date"].fillna("")
)

df[["title", "features"]]
```

```python
cv = CountVectorizer(stop_words="english")

vectors = cv.fit_transform(df["features"])

vectors.shape
```

```python
similarity = cosine_similarity(vectors)

similarity
```

```python
selected_movie_index = 0

scores = list(enumerate(similarity[selected_movie_index]))

scores = sorted(scores, key=lambda x: x[1], reverse=True)

for index, score in scores[1:6]:
    print(df.iloc[index]["title"], "-", score)
```
